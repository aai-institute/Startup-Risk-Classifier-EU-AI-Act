```python
import re
import json
import itertools

# === Step 1: Define correct field tags ===

correct_tags = [
    "AI Use Case",
    "Use Case Description",
    "Risk Classification",
    "Reason",
    "Requires Additional Information",
    "What additional Information"
]

# === Step 2: Auto-generate all variants ===

def generate_tag_variants(tag):
    """Generate all variants by replacing spaces with '', '-' or keeping space."""
    parts = tag.split(" ")
    combinations = list(itertools.product(["", "-", " "], repeat=len(parts)-1))
    variants = set()

    for combo in combinations:
        variant = parts[0]
        for sep, part in zip(combo, parts[1:]):
            variant += sep + part
        variants.add(variant)
    return variants

# Map correct tags to all possible variants
tag_variants = {tag: generate_tag_variants(tag) for tag in correct_tags}

# === Step 3: Normalize malformed headers ===

for correct, variants in tag_variants.items():
    for variant in variants:
        if variant == correct:
            continue  # Don't replace the correct one
        text = re.sub(rf"^{re.escape(variant)}\s*:", f"{correct}:", text, flags=re.MULTILINE)

# === Step 4: Parse text into use case blocks ===

use_case_blocks = re.split(r"########END OF USE CASE########", text.strip())

# === Step 5: Extract data from each block ===

# Pattern to match tags strictly (must end in ":")
split_pattern = re.compile(
    rf"^({'|'.join(re.escape(tag) for tag in correct_tags)}):",
    flags=re.MULTILINE
)

results = []

for block in use_case_blocks:
    block = block.strip()

    if not re.search(r"^AI Use Case:", block, flags=re.MULTILINE):
        continue  # Skip invalid blocks

    chunks = split_pattern.split(block)

    data = {
        "AI Use Case": "",
        "Use Case Description": "",
        "Risk Classification": "",
        "Reason": "",
        "Requires Additional Information": "No",
        "What additional Information": ""
    }

    for i in range(1, len(chunks), 2):
        tag = chunks[i].strip()
        value = chunks[i + 1].strip()
        if tag in data:
            data[tag] = value

    if data["Requires Additional Information"].lower() == "no":
        data["What additional Information"] = ""

    results.append(data)

# === Step 6: Output ===

# print(json.dumps(results, indent=4, ensure_ascii=False))

# Optional: Save to file
with open("ollama_answers.json", "w", encoding="utf-8") as f:
    json.dump(results, f, indent=4, ensure_ascii=False)
```
