```python
import re
import json


# Split the text using the delimiter
use_case_blocks = re.split(r"########END OF USE CASE########", text.strip())

# Define exact expected fields
field_tags = [
    "AI Use Case",
    "Use Case Description",
    "Risk Classification",
    "Reason",
    "Requires Additional Information",
    "What additional Information"
]

# Regex to identify tags (used to split block content)
split_pattern = re.compile(
    rf"^({'|'.join(re.escape(tag) for tag in field_tags)}):",
    flags=re.MULTILINE
)

# Container for results
results = []

for block in use_case_blocks:
    # Remove leading/trailing whitespace
    block = block.strip()

    # Fix: Check if there's an exact "AI Use Case:" field *anywhere* in the block
    if not re.search(r"^AI Use Case:", block, flags=re.MULTILINE):
        continue  # Skip malformed block

    # Split block by tags
    fields = split_pattern.split(block)
    # This gives: ['', 'AI Use Case', 'value', 'Use Case Description', 'value', ...]

    # Prepare output dict
    data = {
        "AI Use Case": "",
        "Use Case Description": "",
        "Risk Classification": "",
        "Reason": "",
        "Requires Additional Information": "No",  # default if missing
        "What additional Information": ""         # default if missing or not needed
    }

    for i in range(1, len(fields), 2):
        tag = fields[i].strip()
        value = fields[i + 1].strip()
        if tag in data:
            data[tag] = value

    # Logic: Remove "What additional Information" if it's not needed
    if data["Requires Additional Information"].lower() == "no":
        data["What additional Information"] = ""

    results.append(data)

# Output as JSON
with open("use_cases.json", "w", encoding="utf-8") as f:
    json.dump(results, f, indent=4, ensure_ascii=False)

# Also print to verify
print(json.dumps(results, indent=4, ensure_ascii=False))
```
